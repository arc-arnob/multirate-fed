```python
import os
import glob

def delete_cluster_files(directory, extension="csv"):
    # Get all files in the directory with the specified extension
    files = glob.glob(f"{directory}/*.{extension}")
    
    # Delete each file
    for file in files:
        os.remove(file)
        print(f"Deleted {file}")
```

```python
delete_cluster_files('../data', extension="csv")
```
